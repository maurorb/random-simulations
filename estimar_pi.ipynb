```python
x = np.linspace(-0.5,0.5,10000)
y = np.linspace(-0.5,0.5,10000)

n=500
Xpontos_sorteados = np.random.choice(x,n)
Ypontos_sorteados = np.random.choice(y,n)

Xpontos_sorteados = np.random.choice(x,n)
Ypontos_sorteados = np.random.choice(y,n)
#estima pi

eval_circle = Xpontos_sorteados**2 + Ypontos_sorteados**2
dentro = (eval_circle<.25).sum()
X_pontos_dentro, Y_pontos_dentro  = Xpontos_sorteados[(eval_circle<.25)], Ypontos_sorteados[(eval_circle<.25)]
X_pontos_fora, Y_pontos_fora  = Xpontos_sorteados[(eval_circle>=.25)], Ypontos_sorteados[(eval_circle>=.25)]
    
fig,ax = plt.subplots(figsize=(9,9))

ax.set_ylim(-0.75,0.75)
ax.set_xlim(-0.75,0.75)

ax.add_patch(
    Circle((0,0),.5,
            fill=False))

ax.add_patch(
    Rectangle((-0.5,-0.5),1,1,
            fill=False))

ax.plot(X_pontos_dentro,Y_pontos_dentro,'b.')
ax.plot(X_pontos_fora,Y_pontos_fora,'r.')

estimativa_pi = 4*dentro/n
print(f'pi ~ {estimativa_pi}')
ax.text(-0.6,-0.6,s = f'pi ~ {estimativa_pi}')
ax.text(-0.6,-0.65,s = f'erro = {round(100*(estimativa_pi- math.pi)/(math.pi),4)} %')
plt.show()
```

# Stress Test

```python
x = np.linspace(-0.5,0.5,10000)
y = np.linspace(-0.5,0.5,10000)
```

```python
n=10000
Xpontos_sorteados = np.random.choice(x,n)
Ypontos_sorteados = np.random.choice(y,n)
```

```python
teste_tamanho = 5000
estimativas = []
erros = []
for n in range(1,teste_tamanho):
    # Sorteia n pontos
    Xpontos_sorteados = np.random.choice(x,n)
    Ypontos_sorteados = np.random.choice(y,n)

    
    #estima pi

    eval_circle = Xpontos_sorteados**2 + Ypontos_sorteados**2
    qtd_dentro = (eval_circle<.25).sum()

    #salva
    est = 4*qtd_dentro/n
    estimativas.append(est)
    erros.append(100*(est - math.pi)/(math.pi))
    
fig,[ax,ax2] = plt.subplots(2,1,figsize=(25,12))

# ax.set_ylim(0,5)
ax.set_xlim(0,teste_tamanho)

ax.plot(range(1,teste_tamanho),estimativas)


#linha verdade
ax.plot([0,teste_tamanho],[math.pi,math.pi],'--',linewidth=1)

ax.set_xlabel('Nmero de pontos escolhidos')
ax.set_ylabel('estimativa para pi')
ax.set_title('Estimativa para pi de acordo com o numero de pontos aleatorios sorteados',fontsize=15)




ax2.set_xlim(0,teste_tamanho)
ax2.plot(range(1,teste_tamanho),erros)
ax2.plot([0,teste_tamanho],[0,0],'--',linewidth=1)


ax2.set_xlabel('Nmero de pontos escolhidos')
ax2.set_ylabel('Erro na estimativa de pi (%)')
ax2.set_title('Erro na estimativa para pi de acordo com o numero de pontos aleatorios sorteados',fontsize=15)

plt.show()
```
